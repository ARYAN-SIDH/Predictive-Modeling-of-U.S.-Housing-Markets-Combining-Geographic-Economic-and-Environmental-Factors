```python
import pandas as pd
import os
import csv
from pathlib import Path
import datetime as dt
```

## LOAD/CLEAN ORIGINAL DATA SOURCE

```python
file_to_load = "./Resources/Sale_Prices_City.csv"
sale_price_df = pd.read_csv(file_to_load, low_memory=False)
sale_price_df
```

```python
clean_sale_df = sale_price_df.dropna()
clean_sale_df = clean_sale_df.drop(columns=['Unnamed: 0'], axis=1)
clean_sale_df
```

```python
# export csv that we'll use to scrape location data
file_path = Path("./Resources/Sale_Prices_City_sample.csv")
clean_sale_df.to_csv(file_path, index=False)
```

### Build framework for properly transposed dataset

```python
# getting ready to transpose data so its indexed by timeframe
# first, drop useless columns
time_df = clean_sale_df.drop(columns=["RegionName", "StateName", "SizeRank"])
time_df
```

```python
# Transpose dataframe to make the date the index, with the RegionID's as the columns
time_transposed = time_df.transpose()
time_transposed.reset_index(drop = False, inplace = True)
time_transposed = time_transposed.rename(columns=time_transposed.iloc[0])
time_transposed = time_transposed.drop([0, 0])
time_transposed = time_transposed.rename(columns={"RegionID": 'DATE'})
time_transposed
```

```python
# convert month column to datetime format
time_transposed['DATE'] = pd.to_datetime(time_transposed['DATE'], infer_datetime_format=True)
# check the data types
time_transposed.dtypes
```

```python
# export transposed dataset to a csv
file_path = Path("./Resources/time_df_transposed.csv")
time_transposed.to_csv(file_path, index=False)
```
